# 1. Import Dependencies

```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import os
import torch._dynamo.utils
```

# 2. Test Environment

```python
env = gym.make("CarRacing-v2", render_mode="human")
```

```python
env.reset()
```

```python
env.render()
```

```python
env.close()
```

```python
# running the environment with random actions
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        action = env.action_space.sample()
        state, reward, done, truncated, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close()
```

# 3. Train Model

```python
env = gym.make("CarRacing-v2", render_mode="human")
env = DummyVecEnv([lambda: env])
```

```python
log_path = os.path.join('C://', 'Users', 'vyshn', 'Documents System', '6th Sem', 'RL', 'RL_Projects', 'Training', 'Logs')
model = PPO('CnnPolicy', env, verbose = 1, tensorboard_log = log_path)
```

```python
model.learn(total_timesteps = 150000)
```

# 4. Save Model

```python
ppo_path = os.path.join('C:\\', 'Users', 'vyshn', 'Documents System', '6th Sem', 'RL', 'RL_Projects', 'Training', 'Saved Models', 'PPO_Driving_Model')
```

```python
model.save(ppo_path)
```

```python
del model
```

```python
model = PPO.load(ppo_path, env)
```

# 5. Evaluate and Test

```python
evaluate_policy(model, env, n_eval_episodes = 10, render = True)
```
